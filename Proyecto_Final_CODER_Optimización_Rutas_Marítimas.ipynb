1. Introducción

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Scraping para obtener datos meteorológicos simulados
url = "https://www.example.com/weather/montevideo"  # URL simulada
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Datos simulados, y orientación de los rumbos
temperatura = "20°C"
viento = "20 nudos"
rumbo_barco = 45  # En grados (0° = Norte, 90° = Este, 180° = Sur, 270° = Oeste)

# Mostrar datos extraídos, agregamos el dato del rumbo propio del barco.
print(f"Temperatura: {temperatura}")
print(f"Velocidad del viento: {viento}")
print(f"Rumbo del barco: {rumbo_barco}°")

# Guardar los datos en un DataFrame se agrega el dato del Rumbo del barco
data = {"Ubicación": ["Montevideo"], "Temperatura": [temperatura], "Velocidad del viento": [viento], "Rumbo": [rumbo_barco]}
df = pd.DataFrame(data)

# Exportar a un archivo CSV
df.to_csv("datos_meteorologicos.csv", index=False)
print("Datos guardados en datos_meteorologicos.csv")

# Simulación de respuesta texto-texto
print("\nRespuesta texto-texto:")
respuesta_texto = """
Para el trayecto de Montevideo a Buenos Aires, se recomienda seguir la ruta principal hacia el norte,
evitar aguas abiertas cercanas a la costa de Mar del Plata debido a mareas altas y vientos fuertes.
Se estima un tiempo de navegación de 30 horas, dado el viento de 20 nudos.
Asegúrese de no llegar a Buenos Aires entre las 10:00 y las 16:00 para evitar restricciones del puerto.
"""
print(respuesta_texto)

# Generar un gráfico con el rumbo del barco
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("Mapa de rutas recomendadas (simulación)", fontsize=14)

# Ruta recomendada (línea verde)
ax.plot([0, 10], [0, 10], label="Ruta recomendada", linestyle='--', color="green")

# Áreas de condiciones adversas
ax.fill_between([4, 6], 4, 6, color="red", alpha=0.5, label="Zona de viento fuerte")
ax.fill_between([7, 8], 7, 8, color="blue", alpha=0.5, label="Zona de mareas altas")

# Representación del rumbo del barco
x_start, y_start = 5, 5  # Punto inicial del barco
dx = np.cos(np.radians(rumbo_barco))  # Desplazamiento en X
dy = np.sin(np.radians(rumbo_barco))  # Desplazamiento en Y
ax.arrow(x_start, y_start, dx, dy, head_width=0.5, head_length=0.5, fc='black', ec='black', label="Rumbo del barco")

ax.set_xlabel("Coordenadas X")
ax.set_ylabel("Coordenadas Y")
ax.legend()
plt.show()
```
